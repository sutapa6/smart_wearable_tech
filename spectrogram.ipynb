```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks')

# give path to data
scope_num = 2
lab_num = 3
filename = f'data/Lab_{lab_num}/scope_{scope_num}.csv'
df = pd.read_csv(filename)

plt.specgram(df['channel_2'], Fs=2, cmap="rainbow",noverlap=250)
  
# Set the title of the plot, xlabel and ylabel
# and display using show() function
plt.title(f"Spectrogram for {filename}")
plt.xlabel("channel_2")
plt.ylabel("TIME")


# save raw scope as png in data/Plots/Lab_<lab_num>/Spectrograms
image_path = f'data/Plots/Lab_{lab_num}/Spectrograms/scope_{scope_num}.png'
plt.savefig(image_path, dpi=300, bbox_inches='tight', transparent=False)
```
